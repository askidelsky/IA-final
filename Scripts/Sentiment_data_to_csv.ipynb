```python
import nltk
import s3fs
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df_baba=pd.read_csv('https://raw.githubusercontent.com/YuanZheng2828/IA-final/main/BABA_5.8.csv')
df_aapl=pd.read_csv('https://raw.githubusercontent.com/YuanZheng2828/IA-final/main/AAPL_5.8.csv')
df_gme=pd.read_csv('https://raw.githubusercontent.com/YuanZheng2828/IA-final/main/GME_5.8.csv')
df_goog=pd.read_csv('https://raw.githubusercontent.com/YuanZheng2828/IA-final/main/GOOG_5.8.csv')
df_tsla=pd.read_csv('https://raw.githubusercontent.com/YuanZheng2828/IA-final/main/TSLA_5.8.csv')
```

```python
def tocsv(df):
    vader=SentimentIntensityAnalyzer()
    f=lambda title:vader.polarity_scores(title)['compound']
    df['compound']=df['title'].apply(f)
    df['date']=pd.to_datetime(df.date).dt.date
    return df
```

```python
df_baba=tocsv(df_baba)
df_aapl=tocsv(df_aapl)
df_gme=tocsv(df_gme)
df_goog=tocsv(df_goog)
df_tsla=tocsv(df_tsla)
```
